```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# ***Standardization***

*Standardization transforms data to have a mean of 0 and a standard deviation of 1.*

It ensures that features with different scales contribute equally to a model's performance.

```python
from sklearn.preprocessing import StandardScaler

# Example dataset
data = pd.DataFrame({
    'Age': [20, 30, 40, 50, 60],
    'Income': [20000, 30000, 40000, 50000, 60000]
})

# Standardization
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data)

# Convert back to DataFrame for readability
standardized_df = pd.DataFrame(standardized_data, columns=['Age', 'Income'])

print("Standardized Data:\n", standardized_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot original data
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data['Age'], bins=10, kde=True, ax=axes[0])
axes[0].set_title('Original Age Distribution')

sns.histplot(standardized_df['Age'], bins=10, kde=True, ax=axes[1])
axes[1].set_title('Standardized Age Distribution')

plt.tight_layout()
plt.show()
```

```python
# Scatter before and after standardization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original data scatter
axes[0].scatter(data['Age'], data['Income'], c='blue', label='Original')
axes[0].set_title("Before Standardization")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Income")

# Standardized data scatter
axes[1].scatter(standardized_df['Age'], standardized_df['Income'], c='green', label='Standardized')
axes[1].set_title("After Standardization")
axes[1].set_xlabel("Standardized Age")
axes[1].set_ylabel("Standardized Income")

plt.tight_layout()
plt.show()
```

# ***Min-Max Normalization***

- Each feature lies between 0 and 1.
- The relative distribution within each feature is preserved.

```python
from sklearn.preprocessing import MinMaxScaler

# Example dataset
data = pd.DataFrame({
    'Age': [20, 30, 40, 50, 60],
    'Income': [20000, 30000, 40000, 50000, 60000]
})

# Min-Max Normalization
scaler = MinMaxScaler(feature_range=(0, 1))
normalized_data = scaler.fit_transform(data)

# Convert back to DataFrame for readability
normalized_df = pd.DataFrame(normalized_data, columns=['Age', 'Income'])

print("Normalized Data:\n", normalized_df)
```

```python


# Plot original data
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data['Income'], bins=10, kde=True, ax=axes[0])
axes[0].set_title('Original Income Distribution')

sns.histplot(normalized_df['Income'], bins=10, kde=True, ax=axes[1])
axes[1].set_title('Normalized Income Distribution')

plt.tight_layout()
plt.show()
```

```python
# Scatter before and after normalization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original data scatter
axes[0].scatter(data['Age'], data['Income'], c='blue', label='Original')
axes[0].set_title("Before Normalization")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Income")

# Normalized data scatter
axes[1].scatter(normalized_df['Age'], normalized_df['Income'], c='green', label='Normalized')
axes[1].set_title("After Normalization")
axes[1].set_xlabel("Normalized Age")
axes[1].set_ylabel("Normalized Income")

plt.tight_layout()
plt.show()
```

# ***Mean Normalization***

Mean Normalization transforms data so that the mean of each feature becomes 0 and its values are scaled within a range of [-1, 1].

```python

import numpy as np

# Example dataset
data = pd.DataFrame({
    'Age': [20, 30, 40, 50, 60],
    'Income': [20000, 30000, 40000, 50000, 60000]
})

# Mean Normalization
mean_normalized_data = (data - data.mean()) / (data.max() - data.min())

print("Mean Normalized Data:\n", mean_normalized_data)
```

```python


# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data['Income'], bins=10, kde=True, ax=axes[0])
axes[0].set_title('Original Income Distribution')

sns.histplot(mean_normalized_data['Income'], bins=10, kde=True, ax=axes[1])
axes[1].set_title('Mean Normalized Income Distribution')

plt.tight_layout()
plt.show()
```

```python
# Scatter before and after normalization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original data scatter
axes[0].scatter(data['Age'], data['Income'], c='blue', label='Original')
axes[0].set_title("Before Normalization")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Income")

# Mean Normalized data scatter
axes[1].scatter(mean_normalized_data['Age'], mean_normalized_data['Income'], c='green', label='Mean Normalized')
axes[1].set_title("After Mean Normalization")
axes[1].set_xlabel("Mean Normalized Age")
axes[1].set_ylabel("Mean Normalized Income")

plt.tight_layout()
plt.show()
```

# ***MaxAbs Scaling***

MaxAbs Scaling scales data based on the maximum absolute value of each feature. It preserves the sign and scales values to the range [-1, 1], ensuring the original distribution and sparsity are maintained.

- It is ideal for datasets with many zeros, as it does not center the data (mean remains unchanged).

```python
from sklearn.preprocessing import MaxAbsScaler

# Example dataset
data = pd.DataFrame({
    'Age': [20, 30, 40, 50, 60],
    'Income': [-20000, 30000, 40000, -50000, 60000]
})

# MaxAbs Scaling
scaler = MaxAbsScaler()
maxabs_scaled_data = scaler.fit_transform(data)

# Convert back to DataFrame for readability
maxabs_scaled_df = pd.DataFrame(maxabs_scaled_data, columns=['Age', 'Income'])

print("MaxAbs Scaled Data:\n", maxabs_scaled_df)
```

```python


# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data['Income'], bins=10, kde=True, ax=axes[0])
axes[0].set_title('Original Income Distribution')

sns.histplot(maxabs_scaled_df['Income'], bins=10, kde=True, ax=axes[1])
axes[1].set_title('MaxAbs Scaled Income Distribution')

plt.tight_layout()
plt.show()
```

```python
# Scatter before and after scaling
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original data scatter
axes[0].scatter(data['Age'], data['Income'], c='blue', label='Original')
axes[0].set_title("Before Scaling")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Income")

# Scaled data scatter
axes[1].scatter(maxabs_scaled_df['Age'], maxabs_scaled_df['Income'], c='green', label='MaxAbs Scaled')
axes[1].set_title("After MaxAbs Scaling")
axes[1].set_xlabel("Age (Scaled)")
axes[1].set_ylabel("Income (Scaled)")

plt.tight_layout()
plt.show()
```

# ***L1 and L2 Normalization***

L1 and L2 normalization are techniques used to rescale data vectors to have specific properties, helping models to converge faster and perform better by ensuring features are comparable.

**Features of L1 and L2 Normalization**

**L1 Normalization**

- **Sum of Absolute Values**: Transforms data such that the sum of absolute values is 1.
- **Sparse Representations**: Useful when sparsity is desired or present in the data.
- **Outlier Sensitivity**: Highly sensitive to outliers.

**L2 Normalization**

- **Sum of Squared Values**: Transforms data so that the Euclidean norm is 1.
- **Balanced Magnitudes**: Helps retain proportional relationships between features.
- **Less Sensitive to Outliers**: Squared terms reduce extreme influence.**bold text**

```python
from sklearn.preprocessing import Normalizer

# Example dataset
data = pd.DataFrame({
    'Feature1': [4, 1, 2, 2],
    'Feature2': [1, 3, 9, 8]
})

# L1 Normalization
l1_normalizer = Normalizer(norm='l1')
l1_normalized = l1_normalizer.fit_transform(data)

# L2 Normalization
l2_normalizer = Normalizer(norm='l2')
l2_normalized = l2_normalizer.fit_transform(data)

# Convert back to DataFrame for readability
l1_normalized_df = pd.DataFrame(l1_normalized, columns=['Feature1', 'Feature2'])
l2_normalized_df = pd.DataFrame(l2_normalized, columns=['Feature1', 'Feature2'])

print("L1 Normalized Data:\n", l1_normalized_df)
print("L2 Normalized Data:\n", l2_normalized_df)
```

```python


# Plot histograms
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Original data histogram
sns.histplot(data['Feature1'], bins=10, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Original Feature1 Distribution')

sns.histplot(data['Feature2'], bins=10, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Original Feature2 Distribution')

# L1 Normalized histogram
sns.histplot(l1_normalized_df['Feature1'], bins=10, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('L1 Normalized Feature1')

sns.histplot(l1_normalized_df['Feature2'], bins=10, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('L1 Normalized Feature2')

plt.tight_layout()
plt.show()
```

```python
# Scatter plots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Original data scatter
axes[0].scatter(data['Feature1'], data['Feature2'], c='blue', label='Original')
axes[0].set_title("Original Data")

# L1 Normalized data scatter
axes[1].scatter(l1_normalized_df['Feature1'], l1_normalized_df['Feature2'], c='green', label='L1 Normalized')
axes[1].set_title("L1 Normalized Data")

# L2 Normalized data scatter
axes[2].scatter(l2_normalized_df['Feature1'], l2_normalized_df['Feature2'], c='orange', label='L2 Normalized')
axes[2].set_title("L2 Normalized Data")

plt.tight_layout()
plt.show()
```

# **`Robust Scaling`**

Robust Scaling transforms features by subtracting the median and scaling according to the interquartile range (IQR). It is robust to outliers as it does not use the mean or standard deviation, which are sensitive to extreme values.

```python
from sklearn.preprocessing import RobustScaler

# Example dataset
data = pd.DataFrame({
    'Feature1': [10, 20, 30, 40, 100],  # Outlier at 100
    'Feature2': [15, 25, 35, 45, 50]
})

# Robust Scaling
scaler = RobustScaler()
robust_scaled_data = scaler.fit_transform(data)

# Convert back to DataFrame for readability
robust_scaled_df = pd.DataFrame(robust_scaled_data, columns=['Feature1', 'Feature2'])

print("Original Data:\n", data)
print("Robust Scaled Data:\n", robust_scaled_df)
```

```python


# Original vs Scaled boxplot
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.boxplot(data=data, ax=axes[0])
axes[0].set_title("Original Data")

sns.boxplot(data=robust_scaled_df, ax=axes[1])
axes[1].set_title("Robust Scaled Data")

plt.tight_layout()
plt.show()
```

```python
# Histogram comparison
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Original data histogram
sns.histplot(data['Feature1'], bins=10, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Original Feature1 Distribution')

sns.histplot(data['Feature2'], bins=10, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Original Feature2 Distribution')

# Scaled data histogram
sns.histplot(robust_scaled_df['Feature1'], bins=10, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Robust Scaled Feature1')

sns.histplot(robust_scaled_df['Feature2'], bins=10, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Robust Scaled Feature2')

plt.tight_layout()
plt.show()
```
