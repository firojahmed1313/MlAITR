# **Working with Axes**

```python
# Create a figure with multiple subplots
plt.subplot(2, 2, 1)  # 2x2 grid, 1st subplot
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Plot 1")

plt.subplot(2, 2, 2)  # 2x2 grid, 2nd subplot
plt.plot([1, 2, 3], [6, 5, 4])
plt.title("Plot 2")

plt.subplot(2, 2, 3)  # 2x2 grid, 3rd subplot
plt.plot([1, 2, 3], [7, 8, 9])
plt.title("Plot 3")

plt.subplot(2, 2, 4)  # 2x2 grid, 4th subplot
plt.plot([1, 2, 3], [9, 8, 7])
plt.title("Plot 4")

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2)

# Plot on each subplot
ax[0, 0].plot([1, 2, 3], [4, 5, 6])
ax[0, 0].set_title("Plot 1")

ax[0, 1].plot([1, 2, 3], [6, 5, 4])
ax[0, 1].set_title("Plot 2")

ax[1, 0].plot([1, 2, 3], [7, 8, 9])
ax[1, 0].set_title("Plot 3")

ax[1, 1].plot([1, 2, 3], [9, 8, 7])
ax[1, 1].set_title("Plot 4")

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
group1 = [3, 7, 8, 5]
group2 = [5, 6, 7, 8]

x = np.arange(len(categories))  # X positions
width = 0.3  # Bar width

# Create Grouped Bar Plot
plt.bar(x - width / 2, group1, width, label='Group 1', color='blue')
plt.bar(x + width / 2, group2, width, label='Group 2', color='orange')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Plot')
plt.xticks(x, categories)  # Set category labels
plt.legend()

plt.show()
```

```python

# Data
categories = ['A', 'B', 'C', 'D']
group1 = [3, 7, 8, 5]
group2 = [5, 6, 7, 8]

# Create Stacked Bar Plot
plt.bar(categories, group1, label='Group 1', color='blue')
plt.bar(categories, group2, bottom=group1, label='Group 2', color='orange')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.legend()

plt.show()
```
