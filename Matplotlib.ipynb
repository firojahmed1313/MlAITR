```python
import matplotlib.pyplot as plt

```

# **Basic Concepts**

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Simple Line Plot')

# Show the plot
plt.show()
```

# **Working with Axes**

```python
# Create a figure with multiple subplots
plt.subplot(2, 2, 1)  # 2x2 grid, 1st subplot
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Plot 1")

plt.subplot(2, 2, 2)  # 2x2 grid, 2nd subplot
plt.plot([1, 2, 3], [6, 5, 4])
plt.title("Plot 2")

plt.subplot(2, 2, 3)  # 2x2 grid, 3rd subplot
plt.plot([1, 2, 3], [7, 8, 9])
plt.title("Plot 3")

plt.subplot(2, 2, 4)  # 2x2 grid, 4th subplot
plt.plot([1, 2, 3], [9, 8, 7])
plt.title("Plot 4")

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2)

# Plot on each subplot
ax[0, 0].plot([1, 2, 3], [4, 5, 6])
ax[0, 0].set_title("Plot 1")

ax[0, 1].plot([1, 2, 3], [6, 5, 4])
ax[0, 1].set_title("Plot 2")

ax[1, 0].plot([1, 2, 3], [7, 8, 9])
ax[1, 0].set_title("Plot 3")

ax[1, 1].plot([1, 2, 3], [9, 8, 7])
ax[1, 1].set_title("Plot 4")

plt.tight_layout()
plt.show()
```

```python
# Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Plot
plt.plot(x, y)

# Customize axes limits
plt.xlim(0, 5)  # X-axis from 0 to 5
plt.ylim(0, 20)  # Y-axis from 0 to 20

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Custom Axes Limits")
plt.show()
```

```python
# Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Plot
plt.plot(x, y)

# Customize ticks
plt.xticks([0, 1, 2, 3, 4], ['Zero', 'One', 'Two', 'Three', 'Four'])
plt.yticks([0, 5, 10, 15, 20])

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Custom Ticks")
plt.show()
```

**Customizing Labels**

```python
# Data
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Plot
plt.plot(x, y)

# Customizing labels
plt.xlabel("Custom X-Axis", fontsize=14, color="blue")
plt.ylabel("Custom Y-Axis", fontsize=14, color="green")
plt.title("Customized Labels", fontsize=18, fontweight="bold")

plt.show()
```

# ***Plot Types***

# ***Line Plot***

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create Line Plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Basic Line Plot')

# Display the plot
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create Line Plot with Customizations
plt.plot(x, y, color='blue', linestyle='--', linewidth=2, marker='o', markersize=8, markerfacecolor='red')

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Customized Line Plot')

plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 3, 4, 5]

# Plot both lines
plt.plot(x, y1, label='Dataset 1', color='blue')
plt.plot(x, y2, label='Dataset 2', color='green')

# Add legend, labels, and title
plt.legend()
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Multiple Lines on the Same Plot')

plt.show()
```

```python

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create Line Plot
plt.plot(x, y, marker='o')

# Add Grid
plt.grid(True, linestyle='--', linewidth=0.5)

# Annotate a Point
plt.annotate('Peak Point', xy=(5, 10), xytext=(3.5, 8),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot with Grid and Annotation')

plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create Line Plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Saved Line Plot')

# Save the plot as an image
plt.savefig('line_plot.png', dpi=300)

plt.show()
```
